```python
#Add your code for this requirement in this cell
import requests

# Define the GitHub and StackOverflow API URLs
github_url = 'https://api.github.com/repos/'
stackoverflow_url_tags = 'https://api.stackexchange.com/2.3/tags/{}/faq?site=stackoverflow'
stackoverflow_url_questions = 'https://api.stackexchange.com/2.3/questions?order=desc&sort=votes&tagged={}&site=stackoverflow'

# Define the topics and repositories to retrieve data for
topics = {
    'git': 'git/git',
    'clone': '',
    'Repository': '',
    'Issue': '',
    'fork': '',
    'pull': '',
    'log': '',
    'merge': '',
    'push': '',
    'Review': '',
    'User Account': '',
    'Organization Account': '',
    'ISSUE_TEMPLATE.md': ''
}

# Function to retrieve data from an API
def get_api_data(url):
    response = requests.get(url)
    data = response.json()
    return data

# Retrieve data for each topic
for topic, repo in topics.items():
    if repo:
        # Retrieve GitHub data
        github_data = get_api_data(github_url + repo)
        print(f"GitHub repository data for {topic}:")
        print(github_data)
    
    # Retrieve Stack Overflow tag data
    stackoverflow_data_tags = get_api_data(stackoverflow_url_tags.format(topic))
    print(f"StackOverflow tag data for {topic}:")
    print(stackoverflow_data_tags)
    
    # Retrieve Stack Overflow question data
    stackoverflow_data_questions = get_api_data(stackoverflow_url_questions.format(topic))
    print(f"StackOverflow question data for {topic}:")
    print(stackoverflow_data_questions)
    
    print("\n")
```

### Requirement 10: 


```python
#Add your code for this requirement in this cell
import requests
from datetime import datetime, timedelta

# GitHub repositories URLs
repos = {
    "angular/angular": "https://github.com/angular/angular",
    "angular/material": "https://github.com/angular/components",
    "angular/angular-cli": "https://github.com/angular/angular-cli",
    "SebastianM/angular-googlemaps": "https://github.com/SebastianM/angular-googlemaps",
    "d3/d3": "https://github.com/d3/d3",
    "facebook/react": "https://github.com/facebook/react",
    "tensorflow/tensorflow": "https://github.com/tensorflow/tensorflow",
    "keras-team/keras": "https://github.com/keras-team/keras",
    "pallets/flask": "https://github.com/pallets/flask"
}

# Stack Overflow API URL
api_url = "https://api.stackexchange.com/2.3"

def get_tags_info(repo_name, repo_url):
    """Get tags information from Stack Overflow API for a repository."""
    try:
        # Get current date and time
        now = datetime.now()

        # Get the date two years ago from now
        two_years_ago = now - timedelta(days=730)

        # Get ISO formatted date strings for the two years ago and current date
        since_date = two_years_ago.strftime('%Y-%m-%d')
        until_date = now.strftime('%Y-%m-%d')

        # Extract owner and repository name from GitHub repository URL
        repo_parts = repo_name.split("/")
        owner = repo_parts[0]
        repo = repo_parts[1]

        # Get tags related to the repository from Stack Overflow API
        tags_url = f"{api_url}/tags"
        params = {
            "order": "desc",
            "sort": "popular",
            "fromdate": since_date,
            "todate": until_date,
            "tagged": f"{owner}/{repo}",
            "site": "stackoverflow"
        }
        response = requests.get(tags_url, params=params)
        tags = response.json()["items"]

        return tags

    except requests.exceptions.RequestException as e:
        print(f"An error occurred while fetching data for {repo_name}: {e}")
        return None

# Retrieve tags information for each repository
for repo_name, repo_url in repos.items():
    print(f"Tags information for repository {repo_name}:")
    tags = get_tags_info(repo_name, repo_url)
    if tags is not None:
        for tag in tags:
            print(f"Tag name: {tag['name']}")
            print(f"Tag count: {tag['count']}")
    print()
```

### Requirement 11: 


```python
import pandas as pd
import requests

# Define the GitHub repository API URLs
repos = [
    "angular/angular",
    "angular/material",
    "angular/angular-cli",
    "SebastianM/angular-googlemaps",
    "d3/d3",
    "facebook/react",
    "tensorflow/tensorflow",
    "keras-team/keras",
    "pallets/flask"
]
url_template = "https://api.github.com/repos/{}/issues?state=all&since={}T00:00:00Z&per_page=100&page={}"

# Define the time range to retrieve the data
start_date = "2021-04-17"
end_date = "2023-04-17"

# Function to retrieve issues data for a given repository
def retrieve_issues_data(repo, start_date, page):
    url = url_template.format(repo, start_date, page)
    response = requests.get(url)
    if response.ok:
        data = response.json()
        if not data:
            return None
        for issue in data:
            issue["repository_url"] = f"https://github.com/{repo}"
        return data
    else:
        print(f"Error retrieving data for {repo}: {response.status_code}")
        return None

# Retrieve the issues data for each repository
issues = []
for repo in repos:
    page = 1
    while True:
        data = retrieve_issues_data(repo, start_date, page)
        if data is None:
            break
        issues.extend(data)
        page += 1

# Create a DataFrame with the issues data
df = pd.json_normalize(issues)

# Check if the "repository_url" column is present in the DataFrame
if "repository_url" not in df.columns:
    print("Error: 'repository_url' column not found in DataFrame")
else:
    # Count the number of issues per repository
    counts = df["repository_url"].value_counts().reset_index()
    counts.columns = ["repository_url", "count"]

    # Get the repository with the highest number of issues
    max_repo = counts.loc[counts["count"].idxmax()]

    # Print the result
    print(f"The repository with the highest number of issues created is {max_repo['repository_url']} with {max_repo['count']} issues.")
```
